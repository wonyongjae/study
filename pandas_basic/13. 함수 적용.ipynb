## 데이터에 함수 적용(apply)

```python
# 키 뒤에 cm을 붙이는 역할

def add_cm(height):
    return str(height) +'cm'

df['키']=df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # pd.notnull = NaN이 아니라면
        return lang.capitalize() # 첫글자만 대문자로
    return lang
df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
