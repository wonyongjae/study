# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키']>=185
```

```python
df[df['키']>=185]
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # filt의 조건을 역으로 이용
```

```python
df[df['키'] >= 185]['수학']
```

```python
df.loc[df['키'] >= 185,'수학']
```

```python
df.loc[df['키'] >=185,['이름','수학','과학']]
```

```python
df.loc[df['키']>=185,['이름','수학','과학']]
```

# 다양한 조건

### &그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

### str함수

```python
filt = df['이름'].str.startswith('송') # 송 씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt]
```

```python
langs = ['Python','Java']
filt = df['SW특기'].isin(langs) # SW특기가 langs 리스트 안에 있는 것과 똑같은 사람들
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
langs = df['SW특기'].str.contains('Java', na=False)
df[langs]
```

```python
df['SW특기'].str.contains('Java', na=True)
```

```python
df['SW특기'].str.contains('Java', na=False)
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
