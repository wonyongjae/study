```python
# matplotlib 라이브러리 불러오기
import matplotlib.pyplot as plt
```

```python
# 현재 실행 화면에 바로 그래프를 그리기
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)  # x축 0~5까지 11개의 데이터 설정
y = x**2
```

**첫번째 그래프 그리기 방법 :  Functional Method**

```python
plt.plot(x, y)  #Fuctional Method (matplotlib 함수)를 불러 그래프를 그리는 방법(x,y데이터를 불러와서 그래프를 그려라)
plt.xlabel('X Label') #X축 레이블 설정
plt.ylabel('Y Label') #Y축 레이블 설정
plt.title('Title');   #그래프 제목 설정
```

Create Multi-plots : 여러개의 그래프를 생성하는 방법

```python
#여러개 그래프 생성(몇개 행, 몇개 열. 몇개 그래프) = 1x 2 영역에 1번째 영역에 그래프 생성
plt.subplot(1, 2, 1)
#그래프 생성 (x값 ,y값, 그래프 색상(r : red))
plt.plot(x,y, 'r')
plt.xlabel('X Label1') #X축 레이블 설정
plt.ylabel('Y Label1') #Y축 레이블 설정
plt.title('Title1');
plt.subplot(1, 2, 2)     #1x 2 영역에 2번째 영역에 그래프생성
plt.plot(x, y, 'b');     #그래프 생성 (x값 ,y값, 그래프 색상(b : blue))
plt.xlabel('X Label2') #X축 레이블 설정
plt.ylabel('Y Label2') #Y축 레이블 설정
plt.title('Title2');
```

**두번째 방법 : Object-oriented Method**

   - 객체(비어있는 캔버스) 생성(크기) 후 축 (x,y) 지정 후 축위에 원하는 그래프 생성 

```python
# create figure object (empty canvas)
fig = plt.figure()

# define the location of the axes
# [ left, bottom, width, height ]
#리스트안에 축에 크기와 위치 지정([그래프 위치(x : 0.1),그래프 위치(y :0.1), 너비(0.8), 높이(0.8)])

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title');

            # x값, y값 그래프 생성
             # plt.xlabel('X Label') => axes.set_xlabel( 'X Label' ) 축에 x축 label 
             # plt.xlabel('Y Label') => axes.set_ylabel( 'Y Label' ) 축에 y축 label 
             # plt.title('Title')  => axes.set_title( 'Set Title' ) 축에 타이틀 생성
```

   - 여러개의 축(axes1, axes2, asex3...) 생성

```python
fig = plt.figure()

#리스트안에 축의 크기와 위치 지정([그래프 위치(x축 0.1),그래프 위치(y축 0.1), 너비(0.8), 높이(0.8)])
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 그래프에 첫번째 축 프레임 생성
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3 ]) # 첫번째 축 프레임 안에 두번째 축 프레임 생성

# 첫번째 축에 x, y 생성, 제목 생성
axes1.plot(x, y)
axes1.set_title('LARGER PLOT') 

# 두번째 축에 x, y 생성, 제목 생성
axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

subplots()

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

Figure Size and DPI

```python
# 그래프 사이즈(figsize) = (가로(8),세로(2)), 해상도(dpi) = 60), dpi = dot per inch
fig = plt.figure(figsize=(8,2), dpi=60)

ax = fig.add_axes([0, 0, 1, 1]) 
ax.plot(x,y);
```

Save a Figure

```python
 #(파일명.확장자명(png, jpg...), 해상도(dpi))
 fig.savefig('my_picture.png', dpi =200)
```

Adding a Legend(범례)

```python
fig = plt.figure()
ax = fig.add_axes([0 ,0 , 1, 1])

ax.plot(x, x ** 2, label = 'X Squared')
ax.plot(x, x ** 3, label = 'X Cubed')

# 범례 설정(범례위치 x,y축 프레임 상대적위 (x,y축 시작점:(0,0), x,y축 끝점(1,1))
ax.legend(loc = (0.5, 0.8));
# 0 ~ 10 locations
```

Plot Appearance

```python
fig = plt.figure() 
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='#FF8C00', linewidth = 2.0, alpha = 1.0,
        linestyle = '-.', marker = '2', markersize=10,
        markerfacecolor = 'b', markeredgewidth= 1,
        markeredgecolor ='red');

# refer to the 'RGB Hex Code' for custom colors
# marker = o, +, *, 1, 2, etc.
```

Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw = 2, ls = '--'  )
#lw = line width, ls = line style
ax.set_xlim([2,3])
ax.set_ylim([5,6])


# x, y축 한계값 또는 특정 x,y축값 위의 데이터 그래프생성(축시작값, 축끝값)  
```
