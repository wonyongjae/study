# 4. 스타일

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'   Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y, linewidth=5);
```

# 마커

```python
plt.plot(x,y, marker='o', linewidth=3)
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

# matplotlib.markers 종류가 많음
공식 홈페이지에 정보 많음

```python
plt.plot(x,y, marker='X', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

# 선 스타일
### matplotlib.linestyle
홈페이지에 정보 많음

```python
plt.plot(x,y,linestyle=':')
plt.plot(x,y,ls=':')
```

```python
plt.plot(x,y,linestyle='--')
plt.plot(x,y,ls='--')
```

```python
plt.plot(x,y,linestyle='-.')
plt.plot(x,y,ls='-.')
```

```python
plt.plot(x,y,linestyle='-')
plt.plot(x,y)
```

### 선 색상

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='#ff0000')
```

```python
plt.plot(x,y,color='b')
```

```python
plt.plot(x,y,c='g')
```

# 포맷

```python
plt.plot(x,y,'ro--') # color, marker, linestyle
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y,'go')
```

```python
plt.plot(x,y,color='green', marker='o', linestyle='None')
```

# 축약어
- markerfacecolor = mfc
- markeredgecolor = mec
- markersize = ms
- linestyle = ls
- linewidth = lw

```python
plt.plot(x,y, marker='o', mfc='r', mec='b', ms='10', ls=':', lw='5')
```

# 투명도

```python
plt.plot(x,y, marker='o', mfc='r', mec='b', ms='10', ls=':', lw='5', alpha=0.7)  # alpha 투명도 [0~1]
```

```python
plt.plot(x,y, marker='o', mfc='r', mec='b', ms='10', ls=':', lw='5', alpha=0.3)  
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=50)  # dots per inch(dpi), 확대, 사이즈는 그대로인채로 조절
plt.plot(x,y)
```

# 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```
