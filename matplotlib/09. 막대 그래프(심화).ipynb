# 9. 막대 그래프(심화)

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'   Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 폰트사용시, 마이너스 표시가 깨지는 현상 해결
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175, 195)
```

# 해치
# matplotlib = Hatch style reference 로 참고

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # XXXX
bar[2].set_hatch('..') #...
```

```python
bar = plt.bar(labels, values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', c='r')
```
