```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # 자동 퍼센트(pct) 소수점 첫번째 자리까지
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C/C++', 'ETC']

explode = [0.2, 0.1, 0, 0, 0, 0]
plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'JavaScript', 'C#', 'C/C++', 'ETC']

#explode = [0.2, 0.1, 0, 0, 0, 0]
explode=[0.05]*6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode, autopct='%.1f%%')
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```
