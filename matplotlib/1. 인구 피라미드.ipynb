# 1. 인구 피라미드

# 남자 데이터 정의

```python
import pandas as pd
file_name = '202109_202109_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.head(3)
```

```python
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int) # 문자열 데이터를 str.replace로 ',' 를 치환한 뒤 타입변경
```

```python
df_m.iloc[0]
```

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # 같은 파일 내에 같은 컬럼이 있기 때문에 .1이 붙어서 식별하게 됨
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)
df_w
```

# 데이터 시각화

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'   Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 폰트사용시, 마이너스 표시가 깨지는 현상 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 1000으로 나누어 "단위: 천 명" 으로 변경
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 위와 동일

plt.title('2021 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png', dpi=100)
```
