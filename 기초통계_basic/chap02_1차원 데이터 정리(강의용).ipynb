## 데이터의 정규화
  - 상대적 결과가 다르므로 통일된 지표로 변환하는 정규화

### 표준화
  
  - 데이터에서 평균을 빼고 표준편차로 나누는 작업
  - 표준화된 데이터는 표준화 변량 혹은 Z 점수

```python
# 시험점수 표준화하기
# (각각 데이터값 - 평균값) / 표준편차
z = (scores1 - np.mean(scores1)) / np.std(scores1)

# 위에서 얻은 값이 표준화이고 표준화된 데이터는 표준화 변량 혹은 Z 점수
z
```

### 도수분포표
   - 분할된 구간과 데이터의 개수를 표로 정리한 것
   - 시험점수 구간을 "계급(class)"
   - 각 계급에 속한 학생수을 "도수(frequency)"
   - 각 구간의 폭을 "계급폭"
   - 계급의 수를 "계급수"

```python
freq, _ = np.histogram(english_scores, bins=10, range=(0, 100))
                                      # bins = 계급수
                                      # range = 계급폭 (최소값, 최대값) 

freq
```

```python
# 도수()
freq
```

```python
# range = 위 값과 같이 최소값 0  최대값 100 그리고 단위 폭 10
freq_class = [f'{i}~{i+10}' for i in range(0, 100, 10)]
freq_dist_df = pd.DataFrame({'frequency':freq},
                            index = pd.Index(freq_class, name='class'))  
                          # i = 소문자, I = 대문자  pd 에서 지원하는 'I'ndex 대문자로 해야됨
freq_dist_df
```

```python
# 0~10, 10~20, ... 이라는 문자열의 리스트를 작성

# freq_class를 인덱스로 DataFrame을 작성
```

```python
# 계급값 (class 값) 구하기
class_value = [(i+(i+10))//2 for i in range(0, 100, 10)]
class_value
```

```python
# 계급값 (class 값) 구하기
```

```python
# 상대도수값 구하기 : 전체 데이터에 대해 해당 계급의 데이터가 어느 정도 비율
rel_freq =  freq / freq.sum() # 해당 계급 데이터 값 / 전체 데이터 값의 합계
rel_freq
```

```python
# 상대도수값 구하기 : 전체 데이터에 대해 해당 계급의 데이터가 어느 정도 비율
```

```python
# 누적상대도수 : 해당 계급까지의 상대도수 합
cum_rel_freq = np.cumsum(rel_freq)
cum_rel_freq
```

```python
# 누적상대도수 : 해당 계급까지의 상대도수 합
```

```python
freq_dist_df['class_value']=class_value
freq_dist_df['relative frequency']=rel_freq
freq_dist_df['cumulative relative frequency']=cum_rel_freq
freq_dist_df = freq_dist_df[['class_value','frequency','relative frequency','cumulative relative frequency']]
freq_dist_df
```

```python
# 계급값과 상대도수와 누적상대도수를 도수분포표에 추가
```
